```python
"""
Example of importing salary data into ipython notebook and making a graph
"""
# imports
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

# import csv
csv_filename = "fall-2015-salary.csv"
# get file into pandas
data = pd.read_csv(csv_filename)

# turn data into pandas dataframe
df = pd.DataFrame(data)

# clean salary numbers
df["Base Salary"] = df["Base Salary"].map(lambda x: float(x.strip('$').replace(',','')))

# make a data frame of just sex and base salary data
df_ss = df[["Sex", "Base Salary"]]

# get the mean base salary per sex
sex_mean = df_ss.groupby("Sex").mean()

# plot base salar vs sex
sex_mean.plot().set_title('Base Salary V. Sex');


# clean annual salary numbers
df["Annualized Salary"] = df["Annualized Salary"].map(lambda x: float(str(x).strip('$').replace(',','')))

# make a data frame of just sex and base salary data
df_sa = df[["Sex", "Annualized Salary"]]

# get the annual base salary per sex
annual_sex_mean = df_sa.groupby("Sex").mean()

# plot annual salary vs sex
annual_sex_mean.plot().set_title('Annualized Salary V. Sex');
```
